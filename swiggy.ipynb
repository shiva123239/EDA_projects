```python
Restaurants_listed_in_a_City=data.groupby('City')['Restaurant'].nunique()
Restaurants_listed_in_a_City
```

```python
plt.figure(figsize=(8,8))
Restaurants_listed_in_a_City.plot(kind='bar',color='#40ff00')
plt.xticks(rotation = 360)
plt.ylabel('Number of Restaurants')
plt.title('Restaurants listed in each City',fontsize=14)
plt.show()
```

```python
Avarage_order_Amount=data.groupby('City')['Price'].mean().sort_values(ascending=False)
Avarage_order_Amount
```

```python
plt.figure(figsize=(9,4))
Avarage_order_Amount.plot(kind = 'bar',color='pink')
plt.xticks(rotation = 360)
plt.ylabel('Avarage Order Price')
plt.title('City Avarage Order Amount',fontsize = 12)
plt.show()
```

```python
Min_delivery_time= data.groupby(['Food type','Restaurant'])['Delivery time'].min().sort_values().head(20)
Min_delivery_time
```

```python
plt.figure(figsize=(9,6))
Min_delivery_time.plot(kind='barh')
plt.xlabel('Time taken to delivery')
plt.show()
```

```python
Max_delivery_time= data.groupby(['Food type','Restaurant'])['Delivery time'].max().sort_values(ascending=False).head(20)
Max_delivery_time

                                 
```

```python
plt.figure(figsize=(9,6))
Max_delivery_time.plot(kind='barh')
plt.xlabel('Time Taken to delivery')
plt.show()
```
